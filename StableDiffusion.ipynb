```python
import torch
from diffusers import StableDiffusionPipeline
import gradio as gr
```

```python
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    use_safetensors=True
).to("cuda")

# ✅ Define image generation function
def generate(prompt):
    image = pipe(prompt, guidance_scale=7.5).images[0]
    return image

# ✅ Build Gradio interface
with gr.Blocks(css="body { background-color: #f9f9f9; }") as demo:
    gr.Markdown("## 🎨 Stable Diffusion ImageBot\nEnter a text prompt and I'll generate an image for you!")

    with gr.Row():
        textbox = gr.Textbox(placeholder="Type a prompt like 'a cyberpunk city at night'...", label="Prompt")
        btn = gr.Button("Generate")

    image_output = gr.Image(label="Generated Image")

    btn.click(fn=generate, inputs=textbox, outputs=image_output)

# ✅ Launch app (you'll get a public link)
demo.launch(share=True)
```
